## Generate simulated data and weights

```python
n = 1250
ns, nt = 7, 25
ssq = 0.1
s = stim.Stim(n, ns, nt)
r = resp.Resp(s, signalType='rank-2', ssq=ssq)
plot.plotFull(s.xy, r.wf)

plt.title('Response')
plt.plot(r.Ysig, 'k', alpha=0.5)
plt.plot(r.Y, 'g', alpha=0.2)
plt.xlabel('trials')
plt.show()
```

## Train and test

### OLS, ARD, Ridge, Lasso

```python
trainPct = 0.8
splitSample = True
fit_these = ['ols', 'ARD', 'ridge', 'lasso']

(Xf0, Y0), (Xf1, Y1) = reg.trainAndTest(s.Xf, r.Y, trainPct=trainPct)
fitters = {'ols': reg.OLS, 'ARD': reg.ARD, 'ridge': reg.Ridge, 'lasso': reg.Lasso}
if not splitSample:
    Xf1 = None
    Y1 = None

# init
fs = {}
for key in fit_these:
    fs[key] = fitters[key](Xf0, Y0, Xf1, Y1, label=key)

# train
for key, obj in fs.iteritems():
    obj.fit()

# test
for key, obj in fs.iteritems():
    obj.predict()
    obj.score()
print
```

### ASD

```python
Ds = s.D
Ds = np.tile(Ds, [s.nt, s.nt])
Ds += 2*np.random.rand(*Ds.shape)
Dt = stim.Dt(s.nt, s.ns)
fs['ASD'] = reg.ASD(Xf0, Y0, Xf1, Y1, Ds=Ds, label='ASD')
fs['ASD'].fit()
fs['ASD'].predict()
fs['ASD'].score()
print
```

### Bilinear

```python
(X0, Y0), (X1, Y1) = reg.trainAndTest(s.X, r.Y, trainPct=trainPct)
fs['bilin'] = reg.Bilinear(X0, Y0, X1, Y1, label='bilin')
fs['bilin'].fit()
fs['bilin'].predict()
fs['bilin'].score()
print
```

## Show true and fitted weights

```python
normalize = True
weights = lambda obj: obj.clf.coef_
label_str = lambda key, wts: '{0} (max={1:0.4f}, mean={2:0.4f}, sum={3:0.4f})'.format(key.upper(), wts.max(), wts.mean(), wts.sum())
vmax = max([weights(obj).max() for obj in fs.values()])

# show true weights
print label_str('TRUE', r.wf)
plot.plotFull(s.xy, r.wf)

# show fitted weights
for key, obj in fs.iteritems():
    wfh = weights(obj)
    print label_str(key, wfh)
    wfh = np.reshape(wfh, [s.nt, s.ns])
    plot.plotFull(s.xy, wfh, vmax=vmax if normalize else None)
```
