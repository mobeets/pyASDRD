## Generate simulated data and weights

```python
n = 500
ns, nt = 7, 25
SNR = 1.0
s = stim.Stim(n, ns, nt)
r = resp.Resp(s, signalType='rank-2', SNR=SNR)
plot.plotFull(s.xy, r.wf)

plt.title('Response')
plt.plot(r.Ysig, 'g', alpha=0.5)
plt.plot(r.Y, 'g', alpha=0.2)
plt.xlabel('trials')
plt.show()
```

```python
fitters = {'ols': reg.OLS, 'ARD': reg.ARD, 'ridge': reg.Ridge, 'lasso': reg.Lasso, 'ASD': reg.ASD}
fits = dict(zip(fitters.keys(), [dict() for i in xrange(len(fitters))]))
for SNR in [0.025, 0.1, 0.25, 0.5, 1.0, 2.0, 5.0, 20.0]:
    for j in xrange(1):
        r = resp.Resp(s, signalType='rank-2', SNR=SNR)
        (X0, Y0), (X1, Y1) = reg.trainAndTest(s.Xf, r.Y, trainPct=0.8)
        (Xs0, Ys0), (Xs1, Ys1) = reg.trainAndTest(s.Xs, r.Y, trainPct=0.8)
        for key, fcn in fitters.iteritems():
            if key == 'ARD':
                continue
            if key == 'ASD':
                obj = fcn(Xs0, Ys0, Xs1, Ys1, Ds=s.D, label='{0} - SNR {1}'.format(key, SNR))
            else:
                obj = fcn(X0, Y0, X1, Y1, label='{0} - SNR {1}'.format(key, SNR))
            obj.fit()
            obj.score(verbose=False)
            if SNR not in fits[key]:
                fits[key][SNR] = []
            fits[key][SNR].append(obj.rsq)
```

```python
colors = {'ols': 'g', 'ridge': 'r', 'lasso': 'b', 'ARD': 'k', 'ASD': 'c'}
for key, vals in fits.iteritems():
    clr = colors[key]
    if key == 'ols':
        continue
    for i, (SNR, vs) in enumerate(vals.iteritems()):
        plt.scatter(SNR, np.array(vs).mean(), c=clr, s=40, alpha=0.7, label=key if i==0 else None)
plt.legend(loc='upper left')
plt.xscale('log')
plt.xlabel('SNR')
plt.ylabel('R-sq (score)')
plt.show()
```
