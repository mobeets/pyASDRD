## Fitting with: OLS

```python
reload(reg)
reload(resp)

# training and testing sets
trainPct = 0.8
(X0, Y0), (X1, Y1) = reg.trainAndTest(S.X, R.Y, trainPct=trainPct)
(Xs0, Y0), (Xs1, Y1) = reg.trainAndTest(S.Xs, R.Y, trainPct=trainPct)
(Xt0, Y0), (Xt1, Y1) = reg.trainAndTest(S.Xt, R.Y, trainPct=trainPct)
(Xf0, Y0), (Xf1, Y1) = reg.trainAndTest(S.Xf, R.Y, trainPct=trainPct)

whs = reg.ols(Xs0, Y0)
wht = reg.ols(Xt0, Y0)
whf = reg.ols(Xf0, Y0)
whtb, whsb = reg.bilinear(X0, Y0)
whb0 = np.outer(wht, whs).flatten()
whb = np.outer(whtb, whsb).flatten()
whfR, _ = reg.ARD(Xf0, Y0)

Yhs = reg.predict(Xs1, whs)
Yht = reg.predict(Xt1, wht)
Yhf = reg.predict(Xf1, whf)
Yhb0 = reg.predict(X1, wht, whs)
Yhb = reg.predict(X1, whtb, whsb)
YhfR = reg.predict(Xf1, whfR)

print 'ols space rmse = {0}'.format(rmse(Y1, Yhs))
print 'ols time rmse = {0}'.format(rmse(Y1, Yht))
print 'ols biln0 rmse = {0}'.format(rmse(Y1, Yhb0))
print 'ols full rmse = {0}'.format(rmse(Y1, Yhf))
print 'als biln rmse = {0}'.format(rmse(Y1, Yhb))
print 'ARD full rmse = {0}'.format(rmse(Y1, YhfR))
```

```python
Xm = S.Xm[-len(Y1):]
plt.scatter(Xm, Y1, 50, c='r', alpha=0.7, label='actual')
plt.scatter(Xm, Yhs, c='g', alpha=0.5, label='ols-s')
plt.scatter(Xm, Yht, c='b', alpha=0.5, label='ols-t')
plt.scatter(Xm, Yhf, c='k', alpha=0.5, label='ols-f')
plt.scatter(Xm, Yhb, c='c', alpha=0.5, label='als-b')
plt.legend(bbox_to_anchor=(1.2, 1.05))
plt.xlabel('stimulus')
plt.ylabel('response')
plt.show()

norm = lambda x: x/np.sum(x)

if R.signalType == 'full':
    ws = np.sum(R.wf, 0)
    wt = np.sum(R.wf, 1)
    wf = R.wf.flatten()
    lbl = 'actual (marginal)'
else:
    ws = R.ws
    wt = R.wt
    wf = np.outer(wt, ws).flatten()
    lbl = 'actual'

    
if R.signalType == 'full':
    plt.plot(norm(wf), '-or', label='actual')
    plt.plot(norm(whf), '-k', label='ols - full')
    plt.plot(norm(whb), '-c', label='als - biln')
    plt.plot(norm(whb0), '--g', label='ols - biln', alpha=0.3)
    plt.legend(bbox_to_anchor=(1.2, 1.05))
    plt.xlabel('normalized weights')
    plt.show()

    plt.plot([min(norm(wf)), max(norm(wf))], [min(norm(wf)), max(norm(wf))], 'r--', lw=3)
    plt.scatter(norm(wf), norm(whf), c='k', label='ols-f')
    plt.scatter(norm(wf), norm(whb), c='c', label='als-b')
    plt.scatter(norm(wf), norm(whb0), c='g', alpha=0.4, label='ols-b')
    plt.legend(bbox_to_anchor=(1.2, 1.05))
    plt.xlabel('actual weights')
    plt.ylabel('fitted weights')
    plt.show()
else:
    plt.plot(norm(ws), '-or', label=lbl)
    plt.plot(norm(whs), '-k', label='ols - space')
    plt.plot(norm(whsb), '-c', label='als - space')
    plt.legend(bbox_to_anchor=(1.2, 1.05))
    plt.xlabel('normalized weights')
    plt.show()

    plt.plot(norm(wt), '-or', label=lbl)
    plt.plot(norm(wht), '-k', label='ols - time')
    plt.plot(norm(whtb), '-c', label='als - time')
    plt.legend(bbox_to_anchor=(1.2, 1.05))
    plt.xlabel('normalized weights')
    plt.show()
```

## Fitting with: ARD, ASD, ASD_FP, ASDRD

```python
wsh = {'ARD': None, 'ASD': None, 'ASD_FP': None, 'ASDRD': None}
wth = {'ARD': None, 'ASD': None, 'ASD_FP': None, 'ASDRD': None}
wfh = {'ARD': None, 'ASD': None, 'ASD_FP': None, 'ASDRD': None}
hys = {'ARD': None, 'ASD': None, 'ASD_FP': None, 'ASDRD': None}
Rgs = {'ARD': None, 'ASD': None, 'ASD_FP': None, 'ASDRD': None}
```

```python
reload(asdard)
fits = [
    'ARD',
    'ASD',
    'ASD_FP',
    'ASDRD'
]

D = S.D

if 'ARD' in fits:
    key = 'ARD'
    wsh[key], Rgs[key], hys[key] = asdard.ARD(Xs0, Y0)
    wth[key], Rgs[key], hys[key] = asdard.ARD(Xt0, Y0)
    wfh[key], Rgs[key], hys[key] = asdard.ARD(Xf0, Y0)
    Yhs = reg.predict(Xs1, wsh[key])
    Yht = reg.predict(Xt1, wth[key])
    Yhf = reg.predict(Xf1, wfh[key])
    print '{0} space rmse: {1}'.format(key, rmse(Yhs, Y1))
    print '{0} time rmse: {1}'.format(key, rmse(Yht, Y1))
    print '{0} biln rmse: {1}'.format(key, rmse(Yhf, Y1))
if 'ASD_FP' in fits:
    key = 'ASD_FP'
    wsh[key], Rgs[key], hys[key] = asdard.ASD_FP(Xs0, Y0, [D])
    Yhs = reg.predict(Xs1, wsh[key])
    print '{0} space rmse: {1}'.format(key, rmse(Yhs, Y1))
#     print '{0}\'s rmse from ols: {1}'.format(key, rmse(wsh[key], whs))
if 'ASD' in fits:
    key = 'ASD'
    wsh[key], Rgs[key], hys[key] = asdard.ASD(Xs0, Y0, [D])
    Yhs = reg.predict(Xs1, wsh[key])
    print '{0} space rmse: {1}'.format(key, rmse(Yhs, Y1))
if 'ASDRD' in fits:
    key = 'ASDRD'
    wsh[key], Rgs[key] = asdard.ASDRD(Xs0, Y0, Rgs['ASD_FP'])
    Yhs = reg.predict(Xs1, wsh[key])
    print '{0} space rmse: {1}'.format(key, rmse(Yhs, Y1))
```
