## Fitting with: OLS

```python
reload(linreg)
reload(resp)

whs = linreg.ols(S.Xs, R.Y)
wht = linreg.ols(S.Xt, R.Y)
whf = linreg.ols(S.Xf, R.Y)
whsb, whtb = linreg.bilinear(S.X, R.Y)
whb0 = np.outer(wht, whs).flatten()
whb = np.outer(whtb, whsb).flatten()

Yhs = resp.predict(S.Xs, whs)
Yht = resp.predict(S.Xt, wht)
Yhf = resp.predict(S.Xf, whf)
Yhb0 = resp.predict(S.X, wht, whs)
Yhb = resp.predict(S.X, whtb, whsb)
```

```python
plt.scatter(S.Xm, R.Y, c='r', alpha=0.5)
plt.scatter(S.Xm, Yhs, c='g', alpha=0.5)
plt.scatter(S.Xm, Yht, c='b', alpha=0.5)
plt.scatter(S.Xm, Yhf, c='k', alpha=0.5)
plt.scatter(S.Xm, Yhb, c='c', alpha=0.1)
plt.show()

norm = lambda x: x/np.sum(x)

if R.signalType == 'full':
    ws = np.sum(R.wf, 0)
    wt = np.sum(R.wf, 1)
    wf = R.wf.flatten()
    lbl = 'actual (marginal)'
else:
    ws = R.ws
    wt = R.wt
    wf = np.outer(wt, ws).flatten()
    lbl = 'actual'

plt.plot(norm(ws), '-or', label=lbl)
plt.plot(norm(whs), '-k', label='ols - space')
plt.plot(norm(whsb), '-c', label='als - space')
plt.legend()
plt.show()

plt.plot(norm(wt), '-or', label=lbl)
plt.plot(norm(wht), '-k', label='ols - time')
plt.plot(norm(whtb), '-c', label='als - time')
plt.legend()
plt.show()

plt.plot(norm(wf), '-or', label='actual')
plt.plot(norm(whf), '-k', label='ols - full')
plt.plot(norm(whb), '-c', label='als - biln')
plt.plot(norm(whb0), '--g', label='ols - biln', alpha=0.2)
plt.legend()
plt.show()
```
