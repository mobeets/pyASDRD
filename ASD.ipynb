```python
import numpy as np
import matplotlib.pyplot as plt
from resp import Resp
from stim import Stim
import linreg
import asdard
```

```python
rmse = lambda x, y: np.sqrt(np.mean((y-x)**2))
```

```python
n, ns, nt = 1000, 19, 7
S = Stim(n, nt, ns)
R = Resp(S)
```

```python
# plt.scatter(np.sum(S.Xb, 1), R.Y, alpha=0.5)
plt.plot(R.ws, 'or')
# plt.plot(R.wt, '-r')
plt.show()
```

```python
plt.plot(np.sum(S.Xb, 1), 'k-', alpha=0.5)
plt.show()
```

```python
plt.plot(R.Y, 'r-', alpha=0.5)
plt.show()
```

```python
whs = linreg.solve(S.Xs, R.Y)
wht = linreg.solve(S.Xt, R.Y)

Yhs = linreg.predict(S.Xs, whs)
Yht = linreg.predict(S.Xt, wht)
```

```python
# plt.scatter(Yhs, R.Y, c='g', alpha=0.5)
# plt.scatter(Yht, R.Y, c='b', alpha=0.5)
plt.scatter(np.sum(S.Xb, 1), R.Y, c='r', alpha=0.5)
plt.scatter(np.sum(S.Xb, 1), Yhs, c='g', alpha=0.5)
plt.scatter(np.sum(S.Xb, 1), Yht, c='b', alpha=0.5)
plt.show()
plt.plot(R.ws, '-k', label='actual')
plt.plot(whs, '-r', label='ols - space')
plt.legend()
plt.show()
```
