```python
rmse = lambda x, y: np.sqrt(np.mean((y-x)**2))
```

## Fitting with: OLS

```python
whs = linreg.solve(S.Xs, R.Y)
wht = linreg.solve(S.Xt, R.Y)
whb = linreg.solve(S.Xb, R.Y)

Yhs = linreg.predict(S.Xs, whs)
Yht = linreg.predict(S.Xt, wht)
Yhb = linreg.predict(S.Xb, whb)
```

```python
plt.scatter(np.sum(S.Xb, 1), R.Y, c='r', alpha=0.5)
plt.scatter(np.sum(S.Xb, 1), Yhs, c='g', alpha=0.5)
plt.scatter(np.sum(S.Xb, 1), Yhb, c='k', alpha=0.1)
plt.scatter(np.sum(S.Xb, 1), Yht, c='b', alpha=0.5)
plt.show()
plt.plot(R.ws, '-or', label='actual')
plt.plot(whs, '-k', label='ols - space')
plt.legend()
plt.show()
```
