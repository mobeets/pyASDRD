```python
X, YA, xy = jake.load('XY.mat')
s = stim.Stim(X, xy)
Ds = s.D
Ds = np.tile(Ds, [s.nt, s.nt])
Dt = stim.Dt(s.nt, s.ns)
fitters = {'ols': reg.OLS, 'ARD': reg.ARD, 'ridge': reg.Ridge, 'lasso': reg.Lasso,
           'ASD': reg.ASD, 'ASD2': reg.ASD, 'ASDRD': reg.ASDRD}
```

```python
for i in xrange(YA.shape[1]):
    Y = YA[:,i]
    (X0, Y0), (X1, Y1) = reg.trainAndTest(s.Xt, Y, trainPct=trainPct)

    fits = {}
    fit_objs = {}
    fit_these = ['ols', 'ridge', 'lasso']
    for key in fit_these:
        fcn = fitters[key]
        obj = fcn(X0, Y0, X1, Y1, label='{0}'.format(key))
        fit_objs[key] = obj
        obj.fit().score(verbose=False)
        fits[key] = obj.rsq
#     print i, fits
```
