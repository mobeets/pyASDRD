```python
Y = YA[:,14]
trainPct = 0.8
fitters = {'ols': reg.OLS, 'ARD': reg.ARD, 'ridge': reg.Ridge, 'lasso': reg.Lasso,
           'ASD': reg.ASD, 'ASD2': reg.ASD, 'ASDRD': reg.ASDRD}
fits = {}
fit_objs = {}
fit_these = ['ols', 'ridge', 'lasso']

Ds = s.D
Ds = np.tile(Ds, [s.nt, s.nt])
Dt = stim.Dt(s.nt, s.ns)

(X0, Y0), (X1, Y1) = reg.trainAndTest(s.Xf, Y, trainPct=trainPct)
(Xs0, Ys0), (Xs1, Ys1) = reg.trainAndTest(s.Xs, Y, trainPct=trainPct)
for key in fit_these:
    fcn = fitters[key]
    if key in ['ASD', 'ASDRD']:
        obj = fcn(X0, Y0, X1, Y1, Ds=Ds, Dt=Dt, label='{0}'.format(key))
    elif key == 'ASD2':
        obj = fcn(Xs0, Ys0, Xs1, Ys1, Ds=Ds, label='{0}'.format(key))
    else:
        obj = fcn(X0, Y0, X1, Y1, label='{0}'.format(key))
    fit_objs[key] = obj
    obj.fit().score(verbose=False)
    fits[key] = obj.rsq
fits
```

```python
reload(plot)
normalize = False
weights = lambda obj: obj.clf.coef_
label_str = lambda key, wts, sc: '{0} (max={1:0.4f}, mean={2:0.4f}, sum={3:0.4f}, rsq={4})'.format(key.upper(), wts.max(), wts.mean(), wts.sum(), sc)
vmax = max([weights(obj).max() for obj in fit_objs.values()])

# show fitted weights
for key in sorted(fit_objs):
    obj = fit_objs[key]
    wfh = weights(obj)
    wfh = np.reshape(wfh, [s.nt, s.ns])
    print label_str(key, wfh, obj.rsq)
    plot.plotFull(s.xy, wfh, vmax=vmax if normalize else None)
```
